```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
```

```python
df = pd.read_csv('../project_code/dataset.csv', header= None)
```

```python
df.shape
```

```python
df.info()
```

```python
data = df.drop(df.loc[(df==0).any(axis=1)])
```

```python
data
```

```python
features = data.iloc[:, :-1].values
labels = data.iloc[:, -1].values
```

```python
# Assuming the labels are strings, convert them to numerical values
label_mapping = {'paper': 0, 'scissors': 1, 'rock': 2}
labels = [label_mapping[label] for label in labels]
```

```python
label = np.array(labels)
```

```python
features.shape
```

```python
label
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(256, activation='relu', input_shape=[features.shape[1],]),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')  
])
```

```python
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
opt = tf.keras.optimizers.Adam(learning_rate=0.001, weight_decay=0.001)

model.compile(loss = loss,
              optimizer = opt,
              metrics=['accuracy'])
```

```python
model.fit(features, label, epochs=300, batch_size=50, validation_split=0.1)
```

```python
model.save('model.h5')
```
